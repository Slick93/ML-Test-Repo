```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
in_data = pd.read_csv("canada_per_capita_income.csv")
in_data
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income in $')
plt.scatter(in_data.year, in_data.capita, color = 'blue', marker = '*')
```

```python
new_in_data = in_data.drop('capita', axis = 'columns')
new_in_data
```

```python
capita = in_data.capita
capita
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_in_data,capita)
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income $')
plt.scatter(in_data.year, in_data.capita, color = 'red', marker = '+')
plt.plot(new_in_data.year, reg.predict(new_in_data), color = 'blue')
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income $')
plt.scatter(in_data.year, in_data.capita, color = 'red', marker = '+')
plt.plot(new_in_data.year, reg.predict(new_in_data[['year']]), color = 'blue')
```
